```python
import os
from pdbfixer import PDBFixer
from openmm.app import PDBFile

input_folder = '/Users/ilayda/Desktop/S4038_structures'
output_folder = '/Users/ilayda/Desktop/fixedd3_pdbs'


os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith('.pdb'):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename.replace('.pdb', '_fixed.pdb'))

        print(f'Fixing {filename}...')

        try:
            fixer = PDBFixer(filename=input_path)
            

            with open(output_path, 'w') as f:
                PDBFile.writeFile(fixer.topology, fixer.positions, f)

            print(f'✓ Saved: {output_path}')
        except Exception as e:
            print(f' Hata: {filename} — {e}')
```
